```python
Q5. What is inheritance? Give an example for each type of inheritance.

Answer : Inheritance is a property of OOPs which helps in inherits the property from one class to another class
         it makes code simple .
         There are 5 types of inheritance in python:
            
        1. Single Level Inheritance - There is only one parent class and one child class.
        Example :
        
        class parent:
            pass
        
        class child(parent):
            pass
        
        2. Multilevel Inheritance - When there is more than one level is present in inheritance then it is called
           multilevel inheritance.
         Example :
            
        class grand_father:
            pass
        
        class father(grand_father):
            pass
        
        class son(father):
            pass
        
        3. Multiple Inheritance - When a child class inherits the property from more than one parent class is 
        called multiple inheritance.
        Example :
        
        class parent_1:
            pass
         
        class parent_2:
            pass
        
        class child(parent_1, parent_2):
            pass
        
        4. Hierarchical Inheritance - When more then one class inherits the property from the same parent class
        is called Hierarchical inheritance.
        Example :
        
        class parent:
            pass
        
        class child_1(parent):
            pass
        
        class child_2(parent):
            pass
        
        5. Hybrid Inheritance - when there is more than one type of inheritance is present then it is called
        Hybrid Inheritance.
        Example :
            
        class A:
            pass
        
        class B(A):
            pass
        
        class C(A):
            pass
        
        class D(B,C):
            pass
```
